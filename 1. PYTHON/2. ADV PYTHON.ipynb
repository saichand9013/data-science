# map reduce filter

```python
#filter() function creates a list of items for which a function returns True:

nums=[3,5,3,5,67,8,3,2,4,6,8]
def is_even(n):
    return n%2==0
evens=list(filter(is_even,nums))          #filter
print(evens)
```

```python
nums=[3,5,3,5,67,8,3,2,4,6,8]
evens=list(filter(lambda n: n%2==0,nums))     #using lambda
print(evens)
```

```python
#map() function applies a function to every item in an iterable:

def update(n):
    return n*2
nums=[3,5,3,5,67,8,3,2,4,6,8]
evens=list(filter(lambda n: n%2==0,nums))
doubled=list(map(update,evens))              #map
print(doubled)
```

```python
nums=[3,5,3,5,67,8,3,2,4,6,8]
evens=list(filter(lambda n: n%2==0,nums))
doubled=list(map(lambda n: n*2,evens))              #map  using lambda
print(doubled)
```

```python
#reduce  we should import the reduce 
from functools import reduce

def addall(a,b):
    return a+b
nums=[3,5,3,5,67,8,3,2,4,6,8]
evens=list(filter(lambda n: n%2==0,nums))
doubled=list(map(lambda n: n*2,evens))
sum=reduce(addall,doubled)                    #reduce
print(sum)
```

```python
nums=[3,5,3,5,67,8,3,2,4,6,8]
evens=list(filter(lambda n: n%2==0,nums))
doubled=list(map(lambda n: n*2,evens))
sum=reduce(lambda a,b: a+b,doubled)                    #reduce using lambda
print(sum)
```

```python
#Decorators let you add extra behavior to a function, without changing the function's code.

def greet(fx):
    def mfx():
        print("good morning")
        fx()
        print("thanks for using this function")
    return mfx
@greet
def hello():
    print("hello world")        

hello()
```

# recursion

```python
#function calling itself

# factorial(7) = 7*6*5*4*3*2*1
# factorial(6) = 6*5*4*3*2*1
# factorial(5) = 5*4*3*2*1
# factorial(4) = 4*3*2*1
# factorial(0) = 1


# factorial(n) = n * factorial(n-1)
def factorial(n):
  if (n == 0 or n == 1):
    return 1
  else:
    return n * factorial(n - 1)


print(factorial(5))
```

```python
print(factorial(10))
print(factorial(12))
```
