# key terms
* datapoints-(2,30)
* x-axis/y-axis- horizontal,vertical line
* marker-symbol to datapoint
* line style-the design of line
* color- colours
* legend- box that tells what represents what
* lable- x and y axis names
* title - title of the plot
* functions-show(),title(),legend(),plot()etccccc

```python
#bar
y=[94,57,87,66,89]
x=["part1","part2","part3","part4","part5"]

plt.bar(x,y)

plt.title('game of thrones')
plt.xlabel('seasons')
plt.ylabel('episodes')

plt.show()
```

```python

y=[94,57,87,66,89]
x=["part1","part2","part3","part4","part5"]
color=["red","green","pink","blue","yellow"]
plt.bar(x,y,color=color)

plt.title('game of thrones')
plt.xlabel('seasons')
plt.ylabel('episodes')

plt.show()
```

```python
# barchart
#plt.bar(x,height, color='',width=value,label='labelname')

product= ['A','B','C','D']
sales=[1000,1500,900,1400]

plt.bar(product,sales,color='orange',label='sales 2025')
plt.show()
```

```python
product= ['A','B','C','D']
sales=[1000,1500,900,1400]

plt.bar(product,sales,color='orange',label='sales 2025')
plt.xlabel('product')
plt.ylabel('sales')
plt.title('complete 2025 data')
plt.legend()


plt.show()
```

```python
product= ['A','B','C','D']
sales=[1000,1500,900,1400]

plt.barh(product,sales,color='orange',label='sales 2025')   #bar h
plt.xlabel('product')
plt.ylabel('sales')
plt.title('complete 2025 data')
plt.legend()


plt.show()  
```

```python
#error bar chart
import numpy as np
x9 = np.arange(0, 4, 0.2)

y9 = np.exp(-x9)

e1 = 0.1 * np.abs(np.random.randn(len(y9)))

plt.errorbar(x9, y9, yerr = e1, fmt = '.-')

plt.show();
```

```python
#line plot
months=[1,2,3,4]
sales=[1111,2222,8638,2534]

plt.plot(months,sales , color='blue' , linestyle=':', linewidth=2, marker='+', ms='20' ,label='2025data')
#plt.plot(we can plot multiple)
plt.title('monthly sales data')
plt.xlabel('months')
plt.ylabel('sales ')
plt.show()
```

```python
#piechart
#plt.pie(values,label=labellist,color=colorlist,autopct='%1.1f%%' to give percentage,

regions=['north','east','west','south']
revenue=[1827,8362,8686,9876]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','green','coral'])
plt.title('states revenue')
plt.show()
```

```python
plt.figure(figsize=(7,7))

x10 = [35, 25, 20, 20]

labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
ex=[0,0,0,1]
plt.pie(x10, labels=labels,explode=ex,shadow = True);

plt.show()
```

```python
plt.figure(figsize=(7,7))

x10 = [35, 25, 20, 20]

labels = ['Computer', 'Electronics', 'Mechanical', 'Chemical']
ex=[0,0,0,1]
plt.pie(x10, labels=labels,explode=ex,shadow = True,startangle=90);

plt.show()
```

```python
#scatter plot
x=np.random.randint(1,10,50)
y=np.random.randint(10,100,50)

plt.scatter(x,y)
plt.show()
```

```python
#box plot
data3 = np.random.randn(100)

plt.boxplot(data3)

plt.show();
```

```python
data3 = np.random.randn(100)
data4 = np.random.randn(100)
data5 = np.random.randn(100)

plot_values=[data3,data4,data5]
plt.boxplot(plot_values)
plt.show()
```

```python
#area plot

# Create some data
x12 = range(1, 6)
y12 = [1, 4, 6, 8, 4]

# Area plot
plt.fill_between(x12, y12)
plt.show()
```

```python
#violin plot
a= [20,65,87,57,9,45,97,654,9,75,345,86,35,8,56]
plt.violinplot(a,showmedians=True)
plt.show()
```

# subplot

```python
x=[1,2,3,4,5]
y=[39,20,48,78,90]

plt.plot(x,y)
plt.show()

x=[6,7,8,9,5]
y=[10,20,39,45,48]

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[39,20,48,78,90]

plt.plot(x,y)
plt.show()
plt.subplot(2,2,1)       #(row,column,position of plot)

x=[6,7,8,9,5]
y=[10,20,39,45,48]
plt.subplot(2,2,2)
plt.plot(x,y)
plt.show()

x=[6,7,8,9,5]
y=[10,20,39,45,48]
plt.subplot(2,2,3)
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[39,20,48,78,90]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('line chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')

plt.show()
```
