```python
#piechart
#plt.pie(values,label=labellist,color=colorlist,autopct='%1.1f%%'

regions=['north','east','west','south']
revenue=[1827,8362,8686,9876]

plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','green','coral'])
plt.title('states revenue')
plt.show()
```
