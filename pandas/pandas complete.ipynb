```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.fillna(0,inplace=True)          #insted of none giving a value 0
df
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
df
```

#Sorting and aggregration

```python
#sorting data in 1 column sort_values()
data={
    "Name":['Ram','Shyam','Ghanesh','Ddhanush'],
    "Age":[28,34,22,30],
    "Salary":[50000,60000,45000,62000],
    "Performance_Score":[85,90,78,92]
}
df=pd.DataFrame(data)
df.sort_values(by="Age",ascending=True, inplace=True)
df
```

```python
data={
    "Name":['Ram','Shyam','Ghanesh','Ddhanush'],
    "Age":[28,34,22,30],
    "Salary":[50000,60000,45000,62000],
    "Performance_Score":[85,90,78,92]
}
df=pd.DataFrame(data)
df.sort_values(by=["Age","Salary"],ascending=[True,False], inplace=True)
df
```

```python
data={
    "Name":['Ram','Shyam','Ghanesh','Ddhanush'],
    "Age":[28,34,22,30],
    "Salary":[50000,60000,45000,62000],
    "Performance_Score":[85,90,78,92]
}
df=pd.DataFrame(data)
meandata=df["Salary"].mean()           #similarly sum(), min() , max() , count() , std()
meandata
```

grouping

```python
data={
    "Name":['Ram','Shyam','Ghanesh','Ddhanush','Aditi','Jagdish','Raj','Simran'],
    "Age":[28,34,22,30,28,34,25,32],
    "Salary":[50000,60000,45000,52000,49000,70000,48000,58000],
    "Performance_Score":[85,90,78,92,88,None,None,89]
}
df=pd.DataFrame(data)
grouped=df.groupby("Age")["Salary"].mean()              #groups the similar age then caluculates the mean
grouped
```
