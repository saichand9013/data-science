```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline              
plt.rcParams['figure.figsize']=6,2           #graph size

import warnings
warnings.filterwarnings('ignore')         #os error

```

```python
vis1=sns.distplot(store["Sales"])
plt.show()
```

```python
vis2=sns.displot(store["Sales"])
plt.show()
```

```python
vis2=sns.displot(store["Sales"],bins=10)
plt.show()
```

```python
vis3=sns.boxplot(data=store,x="totalsale",y="Profit")
plt.show(vis3)
```

```python
vis4=sns.lmplot(data=store,x="totalsale",y="Profit")
plt.show(vis4)
```

```python
vis4=sns.lmplot(data=store,x="totalsale",y="Profit",fit_reg=True)
plt.show(vis4)
```

```python
vis4=sns.lmplot(data=store,x="totalsale",y="Profit",fit_reg=True,hue='Region')
plt.show(vis4)
```
